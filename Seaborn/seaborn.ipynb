# Line Plot

### Line Plot Parameters (sns.lineplot())

x, y : Variables that specify positions on the x and y axes.

hue : Grouping variable that will produce lines with different colors.

size : Grouping variable that will produce lines with different widths. 

style : Grouping variable that will produce lines with different dashes and/or markers.

data : Input data structure.

palettes : Method for choosing the colors to use when mapping the hue semantic.

hue_order : Specify the order of processing and plotting for categorical levels of the hue semantic.

hue_norm : Either a pair of values that set the normalization range in data units or an object that will map from data units into a [0, 1] interval.

sizes : An object that determines how sizes are chosen when size is used.

size_order : Specified order for appearance of the size variable levels, otherwise they are determined from the data. Not relevant when the size variable is numeric.

size_norm : Normalization in data units for scaling plot objects when the size variable is numeric.

markers : Object determining how to draw the markers for different levels of the style variable.

style_order : Specified order for appearance of the style variable levels otherwise they are determined from the data. Not relevant when the style variable is numeric.

estimator : Method for aggregating across multiple observations of the y variable at the same x level. If None, all observations will be drawn.

sort : If True, the data will be sorted by the x and y variables, otherwise lines will connect points in the order they appear in the dataset.

legend : “auto”, “brief”, “full”, or False

```python
a = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
b = [10, 12, 30, 14, 50, 16, 70, 18, 90, 110]
df = pd.DataFrame({'a': a, 'b': b})
df
```

# Histogram

### Parameters of Histogram (sns.distplot())
data : Input data structure. Either a long-form collection of vectors that can be assigned to named variables or a wide-form dataset that will be internally reshaped.

x, y : Variables that specify positions on the x and y axes.

hue : Semantic variable that is mapped to determine the color of plot elements.

weights : If provided, weight the contribution of the corresponding data points towards the count in each bin by these factors.

bins : Generic bin parameter that can be the name of a reference rule, the number of bins, or the breaks of the bins.

binwidth : Width of each bin, overrides bins but can be used with binrange.

cbar_kws : Additional parameters passed to matplotlib.figure.Figure.colorbar().

palette : Method for choosing the colors to use when mapping the hue semantic. 

hue_order : Specify the order of processing and plotting for categorical levels of the hue semantic.

hue_norm : Either a pair of values that set the normalization range in data units or an object that will map from data units into a [0, 1] interval. Usage implies numeric mapping.

color : Single color specification for when hue mapping is not used. Otherwise, the plot will try to hook into the matplotlib property cycle.

legend: If False, suppress the legend for semantic variables.

```python
sns.distplot(dataset['total_bill'])
plt.show()
```

```python
sns.distplot(dataset['total_bill'], bins=50)
plt.show()
```

```python
sns.distplot(dataset['total_bill'], hist=False)
plt.show()
```

```python
sns.distplot(dataset['total_bill'], kde=False)
plt.show()
```

```python
sns.distplot(dataset['total_bill'], color='r')
plt.show()
```

```python
sns.distplot(dataset['total_bill'], vertical=True)
plt.show()
```

# Barplot

### Parameters of Barplot (sns.barplot())
x, y, hue : Inputs for plotting long-form data.

data : Dataset for plotting. If x and y are absent, this is interpreted as wide-form.

order, hue_order : Order to plot the categorical levels in, otherwise the levels are inferred from the data objects.

estimator : Statistical function to estimate within each categorical bin.

ci : Size of confidence intervals to draw around estimated values.

orient : Orientation of the plot (vertical or horizontal).

color : Color for all of the elements.

palette : Colors to use for the different levels of the hue variable.

saturation : Proportion of the original saturation to draw colors at.

errcolor : Color for the lines that represent the confidence interval.

errwidth : Thickness of error bar lines (and caps).

capsize : Width of the “caps” on error bars.

dodge : When hue nesting is used, whether elements should be shifted along the categorical axis.

# Heatmap

### Parameters of Heatmap (sns.heatmap())
data : 2D dataset that can be coerced into an ndarray.

vmin, vmax : Values to anchor the colormap.

cmap : The mapping from data values to color space.

center : The value at which to center the colormap when plotting divergant data.

robust : If True and vmin or vmax are absent, the colormap range is computed with robust quantiles instead of the extreme values.

annot : If True, write the data value in each cell. If an array-like with the same shape as data, then use this to annotate the heatmap instead of the data.

fmt : String formatting code to use when adding annotations.

annot_kws : Properties arguments for annotation.

linewidths : Width of the lines that will divide each cell.

linecolor : Color of the lines that will divide each cell.

cbar : Whether to draw a colorbar.

cbar_kws : Properties arguments for color bar

cbar_ax : Axes in which to draw the colorbar.

square : If True, set the Axes aspect to “equal” so each cell will be square-shaped.

xticklabels, yticklabels : If True, plot the column names of the dataframe. If False, removes the column names.

```python
plt.figure(figsize=(16, 9))
sns.scatterplot(x='who', y='fare', data=dataset, hue='alive', style='alive', size='who', sizes=(50, 200), palette='hot')
plt.show()
```
