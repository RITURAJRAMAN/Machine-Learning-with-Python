# Data Cleaning by deleting Columns and Rows

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing dataset
data = pd.read_csv(r'train.csv')
data
```

```python
# Shape of data
data.shape
```

```python
# Displaying all the columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data.head()
```

```python
# Data Information
data.info()
```

```python
# Data Descriptive statistics
data.describe()
```

```python
# No. of null values in each column
data.isnull().sum()
```

```python
# Plotting Heatmap for null values
plt.figure(figsize=(25, 16))
sns.heatmap(data.isnull())
plt.show()
```

```python
# Checking available percentage of null values
null_var = data.isnull().sum() / data.shape[0] * 100
null_var
```

```python
# Columns having null values more than 17%
null_columns = null_var[null_var > 17].keys()
null_columns
```

```python
# Dropping columns having null values more than 17%
droped_columns = data.drop(columns=null_columns)
droped_columns.shape
```

```python
# Plotting Heatmap for new data after dropping columns having null values more than 17%
plt.figure(figsize=(25, 16))
sns.heatmap(droped_columns.isnull())
plt.show()
```

```python
# Deleting rows having null values
droped_rows = droped_columns.dropna()
droped_rows.shape
```

```python
# Plotting Heatmap for cleaned data
plt.figure(figsize=(25, 16))
sns.heatmap(droped_rows.isnull())
plt.show()
```
