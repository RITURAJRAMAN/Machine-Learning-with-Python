# Data Cleaning of Categorical Data using central tendency (Mode)

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading dataset from Google Drive
datafromdrive = 'https://drive.google.com/file/d/1_Tj6f1AXtKJ5Dj7SJvxGGIN6lKExaom1/view?usp=sharing'
datafromdrive = 'https://drive.google.com/uc?id=' + datafromdrive.split('/')[-2]
data = pd.read_csv(datafromdrive)
data
```

```python
# Checking Data shape
data.shape
```

```python
# Checking data information
data.info()
```

```python
# Data Descriptive statistics
data.describe()
```

```python
# Checking null values
data.isnull().sum()
```

```python
# Plotting heatmap fo null values
plt.figure(figsize=(25, 16))
sns.heatmap(data.isnull())
plt.show()
```

```python
# Selecting categorical data variables for data cleaning
cat_data = data.select_dtypes(include='object')
cat_data
```

```python
# plotting heatmap for null values in categorical variables
plt.figure(figsize=(25, 16))
sns.heatmap(cat_data.isnull())
plt.show()
```

```python
# Checking the percentage of missing values
missing_percent = cat_data.isnull().mean() * 100
missing_percent
```

```python
# getting the columns/variables having null values more than 20%
greater20 = missing_percent[missing_percent > 20].keys()
greater20
```

```python
# Dropping variables having null values more than 20%
cat_data.drop(columns=greater20, axis=1, inplace=True)
```

```python
# Getting variables having null values
missing_percent = cat_data.isnull().mean() * 100
greater0 = missing_percent[missing_percent > 0].keys()
greater0
```

```python
# Filling null values with 'Missing' keyword
cat_data['MasVnrType'].fillna('Missing')
cat_data
```

```python
# Getting the most frequent used value in the 'MasVnrType' variable
cat_data.MasVnrType.mode()
```

```python
# No. of times used in the variable
cat_data.MasVnrType.value_counts()
```

```python
# Filling the null values with most frequent used value in 'MasVnrType' variable
cat_data['MasVnrType'].fillna(cat_data.MasVnrType.mode()[0])
cat_data
```

```python
# Filling all the null values with most frequent used value in respective variables
for var in greater0:
    cat_data[var].fillna(cat_data[var].mode()[0], inplace=True)
cat_data.isnull().sum()
```

```python
# Plotting heatmap for cleaned data
plt.figure(figsize=(25, 16))
sns.heatmap(cat_data.isnull())
plt.show()
```

```python
# Updating data in the original dataset
data.update(cat_data)
data.drop(columns=greater20, axis=1, inplace=True)
data.select_dtypes(include='object').isnull().sum()
```
