```python
df1 = pd.DataFrame({'ID': [1, 2, 3, 4, 5], 'Name': ['R', 'A', 'M', 'A', 'N']})
df2 = pd.DataFrame({'ID': [1, 2, 3, 4, 5], 'Name': ['A', 'B', 'C', 'D', 'E']})
```

```python
# Columns with same labels can't join
df1.join(df2, lsuffix='_1')
```

```python
# Columns with same labels can't join
df1.join(df2, rsuffix='_1')
```
